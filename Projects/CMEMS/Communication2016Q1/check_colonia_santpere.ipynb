```python
import netCDF4
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams.update({'font.size': 20, 'font.family': 'Times New Roman'})
```

```python
datafile = "http://thredds.socib.es/thredds/dodsC/mooring/sea_level/station_coloniasantpere/L1/dep20151028_station-coloniasantpere_L1_latest.nc"
```

```python
with netCDF4.Dataset(datafile) as nc:
    sealevel = nc.variables['SLEV'][:]
    airpressure = nc.variables['AIR_PRE'][:]
    time = nc.variables['time'][:]
    timeunits = nc.variables['time'].units
    dates = netCDF4.num2date(time, timeunits)
```

```python
fig = plt.figure(figsize=(8, 8))
plt.plot(dates, sealevel)
fig.autofmt_xdate()
plt.show()
```
