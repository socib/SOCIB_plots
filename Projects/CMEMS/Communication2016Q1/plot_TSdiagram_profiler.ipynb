```python
import numpy as np
import os
import glob
import netCDF4
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
%matplotlib inline
```

```python
datafile = "http://thredds.priv.socib.es/thredds/dodsC/drifter/profiler_drifter/argo_drifter_apex002-ime_apex002/L1/2012/dep0001_argo-drifter-apex002_ime-apex002_L1_2012-06-09.nc"
```

```python
with netCDF4.Dataset(datafile) as nc:
    lon = nc.variables['LON'][:]
    lat = nc.variables['LAT'][:]
    depth = nc.variables['DEPTH'][:]
    temp = nc.variables['WTR_TEM'][:]
    psal = nc.variables['SALT'][:]
```

```python
fig = plt.figure(figsize=(15,8))
plt.subplot(121)
plt.plot(lon, lat, 'ko')
plt.subplot(122)
plt.plot(temp, psal, 'ko', ms=1)
plt.xlim(12, 30)
plt.ylim(36, 39)
plt.show()
```
