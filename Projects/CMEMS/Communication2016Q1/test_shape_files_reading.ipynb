```python
import shapefile
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
%matplotlib inline
```

```python
worldocean_shape = "/data_local/DataOceano/ShapeFiles/World_Seas.shp"
```

```python
sf = shapefile.Reader(worldocean_shape)
```

```python
sf.shapes
```

```python
fig = plt.figure()
for shape in sf.shapeRecords():
    x = [i[0] for i in shape.shape.points[:]]
    y = [i[1] for i in shape.shape.points[:]]
    plt.plot(x,y)
plt.show()
```
