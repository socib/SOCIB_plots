```python
from skewt import SkewT
import matplotlib as mpl
import os
import glob
%matplotlib inline
mpl.rcParams.update({'font.size': 20})
```

```python
datadir = '/home/ctroupin/Projects/1-Internal/201530627_Rissaga/soundings/'
figdir = '/home/ctroupin/Projects/1-Internal/201530627_Rissaga/figures/soundings/'
```

```python
for datafiles in glob.glob(datadir + 'sounding*.dat'):
    figname = os.path.basename(datafiles) + '.png'
    S = SkewT.Sounding(datafiles)
    S.make_skewt_axes()
    S.add_profile(color='k',bloc=0)
    parcel = S.get_parcel('ml')
    parcel_2=(850.0, 14.6, 14.6, 'user')
    S.lift_parcel(*parcel)
    S.fig.savefig(os.path.join(figdir, figname), dpi=300, facecolor='white', edgecolor='w',
         transparent=True, bbox_inches='tight', pad_inches=0.1)
    S.clear
```
